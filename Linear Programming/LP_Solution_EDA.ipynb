```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from io import StringIO

# Set Seaborn theme to default
sns.set_theme()

# Define the flight assignment data as a multi-line string
data = """
Origin Airport,Destination Airport,2025 Predicted Avg Departure Delay (Minutes),Decision Variable,Total Delay from Route
BWI,CHS,16.01,19,304.19
BWI,LAX,32.58,19,619.02
BWI,LGA,26.64,19,506.16
BWI,MIA,25.15,19,477.85
BWI,OKC,20.17,19,383.23
BWI,SAN,12.9,261,3366.9
BWI,SEA,19.75,19,375.25
MDW,CHS,17.71,40,708.4
MDW,DCA,16.69,173,2887.37
MDW,LAX,23.56,40,942.4
MDW,LGA,15.33,312,4782.96
MDW,MIA,19.41,115,2232.15
MDW,OKC,27.19,40,1087.6
MDW,SAN,21.68,40,867.2
MDW,SEA,22.74,40,909.6
DAL,CHS,15.92,29,461.68
DAL,DCA,17.59,136,2392.24
DAL,LAX,22.98,96,2206.08
DAL,LGA,20.68,29,599.72
DAL,MIA,27.04,29,784.16
DAL,OKC,24.75,29,717.75
DAL,SAN,18.84,29,546.36
DAL,SEA,19.76,207,4090.32
DEN,CHS,14.9,119,1773.1
DEN,LAX,23.05,505,11640.25
DEN,LGA,32.35,42,1358.7
DEN,MIA,25.13,42,1055.46
DEN,OKC,24.25,42,1018.5
DEN,SAN,21.75,42,913.5
DEN,SEA,22.48,42,944.16
LAS,HNL,15.47,60,928.2
LAS,LAX,22.2,71,1576.2
LAS,OKC,17.39,118,2052.02
LAS,SAN,15.96,483,7708.68
LAS,SEA,20.54,39,801.06
"""

# Read the data into a Pandas DataFrame
df = pd.read_csv(StringIO(data))

# Display the first few rows of the DataFrame
print("Flight Assignment Data:")
print(df.head())

# ------------------------------
# 1. Total Delay by Origin Airport
# ------------------------------
# Aggregate total delay per origin airport
delay_by_origin = df.groupby('Origin Airport')['Total Delay from Route'].sum().reset_index()

# Sort the data for better visualization
delay_by_origin = delay_by_origin.sort_values(by='Total Delay from Route', ascending=False)

# Plotting
plt.figure(figsize=(10,6))
sns.barplot(x='Origin Airport', y='Total Delay from Route', data=delay_by_origin)
plt.title('Total Delay by Origin Airport')
plt.xlabel('Origin Airport')
plt.ylabel('Total Delay (Minutes)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ------------------------------
# 2. Total Delay by Destination Airport
# ------------------------------
# Aggregate total delay per destination airport
delay_by_destination = df.groupby('Destination Airport')['Total Delay from Route'].sum().reset_index()

# Sort the data for better visualization
delay_by_destination = delay_by_destination.sort_values(by='Total Delay from Route', ascending=False)

# Plotting
plt.figure(figsize=(10,6))
sns.barplot(x='Destination Airport', y='Total Delay from Route', data=delay_by_destination)
plt.title('Total Delay by Destination Airport')
plt.xlabel('Destination Airport')
plt.ylabel('Total Delay (Minutes)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ------------------------------
# 3. Flight Assignment Mix Across Origin Airports
# ------------------------------
# Aggregate total flight assignments per origin airport
assignments_by_origin = df.groupby('Origin Airport')['Decision Variable'].sum().reset_index()

# Sort the data for better visualization
assignments_by_origin = assignments_by_origin.sort_values(by='Decision Variable', ascending=False)

# Plotting
plt.figure(figsize=(10,6))
sns.barplot(x='Origin Airport', y='Decision Variable', data=assignments_by_origin)
plt.title('Flight Assignment Mix by Origin Airport')
plt.xlabel('Origin Airport')
plt.ylabel('Total Number of Flights Assigned')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ------------------------------
# 4. Flight Assignment Mix by Destination Airport (Stacked Bar Chart)
# ------------------------------
# Pivot the data to get origins as columns and destinations as rows
pivot_destination = df.pivot_table(index='Destination Airport', columns='Origin Airport', values='Decision Variable', fill_value=0)

# Sort the destinations based on total flight assignments
pivot_destination['Total'] = pivot_destination.sum(axis=1)
pivot_destination = pivot_destination.sort_values(by='Total', ascending=False)
pivot_destination = pivot_destination.drop('Total', axis=1)

# Plotting the stacked bar chart
pivot_destination.plot(kind='bar', stacked=True, figsize=(12,8))
plt.title('Flight Assignment Mix by Destination Airport')
plt.xlabel('Destination Airport')
plt.ylabel('Total Number of Flights Assigned')
plt.xticks(rotation=45)
plt.legend(title='Origin Airport', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# ------------------------------
# 5. Heatmap of Flight Assignments Between Origins and Destinations
# ------------------------------
# Create a pivot table for the heatmap
pivot_table = df.pivot(index='Origin Airport', columns='Destination Airport', values='Decision Variable').fillna(0)

# Plotting the heatmap
plt.figure(figsize=(12,8))
sns.heatmap(pivot_table, annot=True, fmt=".0f")
plt.title('Heatmap of Flight Assignments Between Origins and Destinations')
plt.xlabel('Destination Airport')
plt.ylabel('Origin Airport')
plt.tight_layout()
plt.show()
```
